### 1.1 Image creation

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).


a) Create and display a grayscale image that has 150 rows and 200 columns where each value should be of type unsigned integer with 8 bits. 

```python
#First we need to initialize a new matrix specifying its dimensions (200 x 150)
image1 = np.zeros([150,200], dtype = 'uint8')

#Change the values of the matrix in order to see a different types of color gray
image1[:37,:] = 25
image1[37:74,:] = 50
image1[74:111,:] = 75
image1[111:,:] = 100

#We need specify the values of cmap, vmin and vmax to display a grayscale image
plt.imshow(image1, cmap= 'gray',vmin=0, vmax=255)
plt.show()
```

How many values can each pixel have in this case? 

**Solution:** 

This allows us to represent 2^8 = 256 values in each pixel.

### 1.4 Images manipulation

a) Add a black frame to Lenna's image. It should have a width of 1/10th of the height of the original image. We recommend to create a bigger black image to later embed the original image in the central part.

```python
width = lennaImage.shape[0] * 0.1 #First we need to calculate the width of the frame
width = np.ceil(width).astype(int) #We need to convert this number into a round integer number to 
                                   #get the shape of the frame
    
frame = np.zeros(shape=(lennaImage.shape[0]+2*width,lennaImage.shape[1]+2*width,3),dtype=np.uint8) 
#Convert also the type of the matrix (from float64 to uint8) because we want to "insert" lennaImage later...
```

```python
frame[width:-width,width:-width,:] = lennaImage #Now we can insert the lennaImage into the black frame taking 
#again the frame width as the limit

#Display the image...
plt.imshow(frame,cmap='gray')
plt.show()
```

b) Change the mask of the framed image to red. Visualize the three images with their respective titles.

```python
#Now we'll go for the red one...
red_frame = np.zeros(shape=(lennaImage.shape[0]+2*width,lennaImage.shape[1]+2*width,3),dtype=np.uint8)
#We need to set red_frame's "first channel" (Red channel) all to its maximum value -> 255
red_frame[:,:] = [255,0,0]

#Insert lenna's image...
red_frame[width:-width,width:-width,:] = lennaImage

#Now we display the image with the red frame...
plt.imshow(red_frame)
plt.show()
```

```python
titles = ["Lenna's original image", "Lenna's image with black frame", "Lenna's image with red frame"] #First we need the titles...
images = [lennaImage, frame, red_frame] #and then the images...

fig, axs = plt.subplots(nrows=2, ncols=2)

#original image
ax_rgb = axs[0,0]
ax_rgb.axis('off')
ax_rgb.set_title(titles[0])
ax_rgb.imshow(images[0])

#gray image
ax_gray = axs[0,1]
ax_gray.axis('off')
ax_gray.set_title(titles[1])
ax_gray.imshow(images[1], cmap='gray')

#binary image
ax_gray = axs[1,0]
ax_gray.axis('off')
ax_gray.set_title(titles[2])
ax_gray.imshow(images[2], cmap='gray')

#empty image
axs[1,1].axis('off')

plt.show()
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through scikit-image command. Check the minimal and maximal values of the enhanced image, and compare them with the previous one. 

Implement the contrast enhancement using the two different available methods.

```python
#METHOD 1 - Manual enhancedment function applying the histogram manipulation
def manual_enhancement(image,min_val,max_val,range_val):
    return ((image-min_val)/(max_val-min_val))*range_val
```

```python
car2 = io.imread('images/car_lowContrast.jpg')
#We will use the function above...
car2 = manual_enhancement(car2,np.min(car2),np.max(car2),255)

#Finally we need to display the image...
plt.imshow(car2,cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
#METHOD 2 - exposure.rescale_intensity function from skimage
from skimage import exposure

car3 = io.imread('images/car_lowContrast.jpg') #Get the image...

car3 = exposure.rescale_intensity(car3, in_range=(63,193)) #Only by applying the function we obtain the 
#rescaled image with contrast enhancement... 

#Finally display the image...
plt.imshow(car3,cmap='gray',vmin=0,vmax=255)
plt.show()
```

### 1.6 Masks and logical image manipulation

Read the image `circles.bmp` (see figure top) and define 3 images in such a way that each one contains only 1 ball (see figure bottom). Visualize the 3 images in the same figure, each one with its corresponding title.

<img src="images_notebook/ex16.png">

**Note:** Be careful when applying `imshow` since it automatically increases the contrast of the image (putting minimum values to 0 and maximal to 1).

**Note:** Be careful when applying type conversion. Check the ranges of the image values.

```python
im = io.imread('images/circles.bmp') #Load the image...

new_im = np.ones(shape=im.shape,dtype=np.uint8) * 255 #We create an auxiliar fully white image 

mask = im < im.max() #We create a mask whith all 3 circles
first_circle = im==im.min() #Mask for the black cirle
last_circle = im==im[mask].max() #Mask for the light gray circle
middle_circle = (im>im.min())*(im<im[mask].max()) #Mask for the gray circle

#Then we use the previous masks in order to generate the new images...
im_first = new_im.copy()
im_second = new_im.copy()
im_last = new_im.copy()

im_first[first_circle] = im[first_circle] #Using the first mask an so on...
im_second[middle_circle] = im[middle_circle]
im_last[last_circle] = im[last_circle]

fig, axs = plt.subplots(nrows=2, ncols=2)

#original image
ax_original = axs[0,0]
ax_original.axis('off')
ax_original.set_title("Original image")
ax_original.imshow(im, vmin=0, vmax=255)

#direct manipulation image
ax_first = axs[0,1]
ax_first.axis('off')
ax_first.set_title("Left circle")
ax_first.imshow(im_first, vmin=0, vmax=255)

#command manipulation image
ax_second = axs[1,0]
ax_second.axis('off')
ax_second.set_title('Middle circle')
ax_second.imshow(im_second, vmin=0, vmax=255)

#empty image
ax_last = axs[1,1]
ax_last.axis('off')
ax_last.set_title('Right circle')
ax_last.imshow(im_last, vmin=0, vmax=255)

plt.show()
```

### 1.7 Color manipulation

a) Create and visualize the three images shown in the following figure:
<img src="./images_notebook/ex17a.png">

```python
#First we're going to create the left square...
redIm = np.zeros(shape=(200,200),dtype=np.uint8)#Initialize a 200 x 200 matrix with 0s (all black)
redIm[:,100:] = 255#Then we'll take all of its rows and we'll set all the right part(from 100 to 200 column) to 255(white)

#Then we only need to do the same for the other two...
greenIm = np.zeros(shape=(200,200),dtype=np.uint8)
greenIm[100:,:] = 255
blueIm = np.zeros(shape=(200,200),dtype=np.uint8)
blueIm[:100,:100] = 255

#Last we're going to visualize the three images
plt.imshow(redIm,cmap='gray',vmin=0,vmax=255)
plt.show()
plt.imshow(greenIm,cmap='gray',vmin=0,vmax=255)
plt.show()
plt.imshow(blueIm,cmap='gray',vmin=0,vmax=255)
plt.show()
```

b) Using these three images, construct the following color image:
<img src='./images_notebook/ex17b.png'>

```python
colored_Image = np.zeros(shape=(200,200,3),dtype=np.uint8)#To create the coloured image we will need another 
#matrix of 200 x 200 and with 3 'channels'

#We will put in order the first image in the red channel, the second one is going to be the green channel and so on...
colored_Image[:,:,0] = redIm 
colored_Image[:,:,1] = greenIm
colored_Image[:,:,2] = blueIm

#We can see we composed a new coloured image by the three we made initially
plt.imshow(colored_Image)
plt.show()
```
