### 1.1 Image creation

Remember that [Numpy](https://docs.scipy.org/doc/numpy-1.13.0/reference/) allows us to define images of different kinds treating them as matrices.

The most common examples are:

* Binary: with values of 0 or 1, interpreted as black or white, respectively.
* Grayscale: with possible data types uint8, uint16, double, whose pixel values define their intensity.
* RGB Image: color images consisting of 3 matrices (usually called 'channels'), which contain intensity values for each color separately R (red), G (green) and B (blue).


a) Create and display a grayscale image that has 150 rows and 200 columns where each value should be of type unsigned integer with 8 bits. 

```python
#First we need to initialize a new matrix specifying its dimensions (200 x 150)
image1 = np.zeros([150,200], dtype = 'uint8')

#Change the values of the matrix in order to see a different types of color gray
image1[:37,:] = 25
image1[37:74,:] = 50
image1[74:111,:] = 75
image1[111:,:] = 100

#We need specify the values of cmap, vmin and vmax to display a grayscale image
plt.imshow(image1, cmap= 'gray',vmin=0, vmax=255)
plt.show()
```

How many values can each pixel have in this case? 

**Solution:** 

This allows us to represent 2^8 = 256 values in each pixel.

b) Change the color of the pixels of the image in such a way that the left half is black (0) and the right half is white (255). 

Visualize the image and print the values of its first row.

```python
#First we create a black image (with all its values equal to 0) with dimension 200 x 150
image2 = np.zeros([150,200], dtype = 'uint8')

#Change the values in order to see the left half image black and the other half white
image2[:,100:] = 255

#Display the picture
plt.imshow(image2, cmap= 'gray',vmin=0, vmax=255)
plt.show()
```

### 1.3 Color image manipulation

a) Load and show the image `lenna.jpg`, and check its dimensions.

**Note:** Pictures may appear in the figure with pseudocolors. Use (imread,im,cmap='rgb') to display them in RGB color and (imread,im,cmap='gray') to display them in grayscale. For more information, check the color maps available in the documentation.

```python
lennaImage = io.imread('images/lenna.jpg', cmap='rgb') #Load the image in rgb mode...

#Then we show the image and its dimensions...
plt.imshow(lennaImage)
plt.show()
print("Its dimensions are: ", lennaImage.shape[1], " x ", lennaImage.shape[0] )
```

b) Convert the previous image to grayscale, visualize its. and save it. Moreover, show both images (RGB & greyscale) details such as shape, dtype, max and min values.

```python
"""
ATENTION: Repair warning!
"""
from skimage.color import rgb2gray

lennaGray = rgb2gray(lennaImage) #We need to convert the previous image from rgb to grayscale with rgb2gray function

#Display the image
plt.imshow(lennaGray, cmap = 'gray')
plt.show()

#Finally we save the image using io.imsave
io.imsave('lenna_grayscale.jpg', lennaGray)
```

```python
#First we display the rgb image with its properties below
plt.imshow(lennaImage)
plt.show()
print("Shape: ", lennaImage.shape[1], " x ", lennaImage.shape[0], ", dtype: ", lennaImage.dtype, ", max value: ", lennaImage.max(), ", min value: ", lennaImage.min())

#At least the greyscale image with its properties below
plt.imshow(lennaGray, cmap = 'gray')
plt.show()
print("Shape: ", lennaGray.shape[1], " x ", lennaGray.shape[0], ", dtype: ", lennaGray.dtype, ", max value: ", lennaGray.max(), ", min value: ", lennaGray.min())
```

c) Convert the lena image into a boolean image, by applying a binarization to the grayscale image. This should be done by setting to 0 all values equal or below 100 and to 1 otherwise. Check the range of the values of the image, and print the 10th row.

```python
#First we reload the lenna.jpg into 'im' variable
im = io.imread('images/lenna.jpg')

#There is another way to convert the rgb image to grayscale image...
#The method consist in applying the mean of the three matrix (each one with its particular rgb values) 
im = im.mean(axis=2)
plt.imshow(im,cmap='gray')
plt.show()

#Once we obtain the grayscale matrix lets check its values in order to apply binarization
bin_im = (im > 100)

#'mod_im' is the matrix with 0s and 1s created from 'bin_im' that we're going to display
mod_im = np.zeros(shape=bin_im.shape)
mod_im[bin_im] = 1

#Display the image ant its 10th row
plt.imshow(mod_im, cmap='gray')
plt.show()

print("The 10th row of the matrix contain the following values: \n\n", mod_im[10,:])
```

d) Visualize in the same figure three images: (1) the original image, (2) the graylevel image, and (3) the binarized image. Set the corresponding titles of the images.

```python
titles = ["Lenna's original image", "Lenna's grayscale image", "Lenna's binari image"] #First we need the titles...
images = [lennaImage, lennaGray, mod_im] #and then the images...

for i in range(3): #we have 3 images
    plt.subplot(2,2,i+1) #we're doing a grid in order to display 3 images at the same time
    plt.title(titles[i]) #display titles
   
    #display images...
    if i == 0:
        plt.imshow(images[i])
    else:
        plt.imshow(images[i],cmap='gray')
        
    #remove axis values
    plt.xticks([])
    plt.yticks([])
    
plt.show()#finally display the entire grid
```

### 1.4 Images manipulation

a) Add a black frame to Lenna's image. It should have a width of 1/10th of the height of the original image. We recommend to create a bigger black image to later embed the original image in the central part.

```python
width = lennaImage.shape[0] * 0.1
width = np.ceil(width).astype(int)

frame = np.zeros(shape=(lennaImage.shape[0]+2*width,lennaImage.shape[1]+2*width,3),dtype=np.uint8) #Matrix that contains np.float64
print(frame.shape)
print(type(frame[0,0,0]))

print(lennaImage.shape)
print(type(lennaImage[0,0,0]))

plt.imshow(frame)
plt.show()
```

```python
frame[width:-width,width:-width,:] = lennaImage
plt.imshow(frame,cmap='gray')
plt.show()
```

b) Change the mask of the framed image to red. Visualize the three images with their respective titles.

```python
red_frame = np.zeros(shape=(lennaImage.shape[0]+2*width,lennaImage.shape[1]+2*width,3),dtype=np.uint8)
red_frame[:,:,0] = 255
red_frame[width:-width,width:-width,:] = lennaImage
plt.imshow(red_frame)
plt.show()
```

```python
titles = ["Lenna's original image", "Lenna's image with black frame", "Lenna's image with red frame"] #First we need the titles...
images = [lennaImage, frame, red_frame] #and then the images...

for i in range(3): #we have 3 images
    plt.subplot(2,2,i+1) #we're doing a grid in order to display 3 images at the same time
    plt.title(titles[i]) #display titles
   
    #display images...
    plt.imshow(images[i])
        
    #remove axis values
    plt.xticks([])
    plt.yticks([])
    
plt.show()#finally display the entire grid
```

### 1.5 Contrast enhancement

a) Open the image `car_lowContrast.jpg` and visualize it without graylevel enhancement. To this purpose, check which are the minimal and maximal values and how to use the command imshow without contrast enhancement.

Note: When running the function `imshow()` the image is authomatically enhanced. Use the `vmin` and `vmax` parameters in order to deactivate the enhancement applied by default.

```python
car = io.imread('images/car_lowContrast.jpg')
print(np.min(car),np.max(car))
plt.imshow(car,cmap='gray',vmin=0,vmax=255)
plt.show()
```

b) Create an image called `dark_image` using the  values from the original one, so that the minimal value of the new image is 0. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
dark_image = car.copy()
dark_image -= 63
print(np.min(dark_image),np.max(dark_image))
plt.imshow(dark_image,cmap='gray',vmin=0,vmax=255)
plt.show()
```

c) Create a new image called `bright_im` from the original one, so that the maximal value of the new image is 255. Check the minimal and maximal values of the new image, and compare them with the previous one.

```python
bright_im = car.copy()
bright_im += 255 - 193
print(np.min(bright_im),np.max(bright_im))
plt.imshow(bright_im,cmap='gray',vmin=0,vmax=255)
plt.show()
```

d) Apply contrast enhancement to the original image. There are two ways for it: (1) by direct manipulation, and (2) through scikit-image command. Check the minimal and maximal values of the enhanced image, and compare them with the previous one. 

Implement the contrast enhancement using the two different available methods.

```python
car2 = car.copy()
print(np.min(car),np.max(car))
plt.imshow(car2,cmap='gray')
plt.show()
```

```python
from skimage import exposure

car3 = car.copy()
img_rescale = exposure.rescale_intensity(car3, in_range=(63,193))

# Ecualización
img_eq = exposure.equalize_hist(car3)

# Ecualización adaptiva
img_adapteq = exposure.equalize_adapthist(car3, clip_limit=0.03)

print(np.min(img_eq),np.max(img_eq))
for eq in (img_eq, img_adapteq):
    plt.imshow(eq,cmap='gray')
    plt.show()
```
